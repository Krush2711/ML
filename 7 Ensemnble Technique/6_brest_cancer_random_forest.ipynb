```python
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
```

```python
df = load_breast_cancer()
```

```python
print(df.DESCR)
```

```python
x = df.data
y = df.target
```

```python
x_t, x_tt, y_t, y_tt = train_test_split(x, y, test_size=.2 , random_state=42)
```

```python
rf = RandomForestClassifier(random_state=42)
rf.fit(x_t, y_t)
```

```python
rf.score(x_tt, y_tt)
```

```python
rf2 = RandomForestClassifier(
    n_estimators=100, 
    criterion='log_loss',
    max_depth=4,
    min_samples_split=2,
    min_samples_leaf=1,
    max_features='sqrt',
    bootstrap=True, 
    oob_score=True,
    n_jobs=-1,
    random_state=42
)
```

```python
rf2.fit(x_t, y_t)
```

```python
rf2.score(x_tt, y_tt)
```
