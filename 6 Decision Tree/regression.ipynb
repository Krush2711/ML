```python
import numpy as np 
```

```python
import matplotlib.pyplot as plt 
from sklearn.tree import DecisionTreeRegressor
```

```python
np.random.seed(42)
x = np.random.rand(200, 1) - 0.5 
y = x ** 2 + 0.025 * np.random.rand(200, 1)
```

```python
plt.scatter(x, y, c='r')
```

```python
import dtreeviz
random_state = 1234 
```

```python
DTR = DecisionTreeRegressor(max_depth=2, random_state=42)
DTR.fit(x,y)
```

```python
viz_rmodel = dtreeviz.model(model=DTR, 
                            X_train=x, 
                            y_train=y, 
                            )
```

```python
viz_rmodel.view(scale=2.5)
```

```python
from sklearn.tree import plot_tree
plot_tree(DTR, filled=True, rounded=True )
```
