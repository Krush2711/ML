```python
import numpy as np
```

```python
x = np.array([1, 2, 3])
y = np.array([1 ,2 ,2])
n = len(x)
```

Pramamter 

```python
m = 0 
b = 0
lr = 0.1 #learing rate
epochs = 100
```

Gradient Descent loop


```python
for epoc in range(epochs):
    y_pred = m *x + b
    err = y_pred - y 
    # print(err)
    
    dm = (2/n) * np.sum(err * x)
    db = (2/n) * np.sum(err)
    
    m = m - lr*dm
    b = b - lr*db
```

# Multi Dimension

```python
X = np.array([
    [1, 1],
    [1, 2],
    [1, 3]
])  # first column = bias

y = np.array([1, 2, 2])
n = len(y)
```

```python
theta = np.zeros(X.shape[1])
lr = 0.1 #learing rate
epochs = 1000
```

```python
for epoc in range(epochs):
    y_pred = X @ theta 
    err = y_pred - y 
    # print(err)
    
    gradient = (2/n) * (X.T @ err)
    
    theta = theta - lr * gradient

print(theta)
```
