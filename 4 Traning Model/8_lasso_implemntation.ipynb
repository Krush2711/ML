```python
class MyLassoreGD:
    def __init__(self, lr = 0.01, epochs=1000, alpha =0.1):
        """
        lr : leraing rate 
        epochs : number of itration 
        alpha : lambda (rehulariation)
        """
        
        self.lr = lr 
        self.epochs = epochs
        self.alpha = alpha
        
    def fit(self , x, y):
        n_sample, n_features = x.shape
        x_b = np.c_[np.ones(n_sample), x]
        
        self.tehta_ = np.zeros(n_features + 1)
        
        for _ in range(self.epochs):
            y_pred = x_b @ self.tehta_
            error = y_pred -y 
            grad = (2/ n_sample) * (x_b.T @ error)
            grad[1:] += self.alpha * np.sign(self.tehta_[1:])
            self.tehta_ -= self.lr * grad
            
            
        self.intercept_ = self.tehta_[0]
        self.coef_ = self.tehta_[1:]
        
        return self

    def predict(self, x):
        x_b = np.c_[np.ones(x.shape[0]), x]
        return x_b @ self.tehta_
```

```python
# Dummy data
X = np.array([[1, 2],
              [2, 1],
              [3, 0],
              [4, 1]])
y = np.array([5, 6, 7, 10])

# Scale features (VERY IMPORTANT for Lasso)
X = (X - X.mean(axis=0)) / X.std(axis=0)

model = MyLassoreGD(alpha=0.5, lr=0.05, epochs=2000)
model.fit(X, y)

print("Intercept:", model.intercept_)
print("Weights:", model.coef_)
```
