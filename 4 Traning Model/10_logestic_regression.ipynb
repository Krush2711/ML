```python
import numpy as np 
```

```python
class  MyLogestic:
    def __init__(self, lr = 0.1, epochs = 1000):
        self.lr= lr
        self.epochs = epochs
        
    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
    
    def fit(self, x, y):
        n_sample , n_features = x.shape
        
        self.w = np.zeros(n_features)
        self.b = 0.0
        
        for _ in range(self.epochs):
            
            z = x @ self.w + self.b
            
            # probabity prediction
            p = self.sigmoid(z)
            
            # gradianet
            
            dw = (1 / n_sample) * (x.T @ (p -y))
            db = (1 / n_sample) * np.sum(p - y)
            
            # update 
            self.coef_ = self.w
            self.intercept = self.b 
            return self 
    
    def pred_proabablity(self, x):
        z = x @ self.w + self.b
        return self.sigmoid(z)
    
    def predict(self, X, threshold=0.5):
        probs = self.pred_proabablity(X)
        return (probs >= threshold).astype(int)
        
```

```python
# Example dataset
X = np.array([[1, 2],
              [2, 3],
              [3, 4],
              [4, 5]])
y = np.array([0, 0, 1, 1])

# Feature scaling (VERY IMPORTANT)
X = (X - X.mean(axis=0)) / X.std(axis=0)

model = MyLogestic(lr=0.1, epochs=2000)
model.fit(X, y)

print("Weights:", model.coef_)
print("Bias:", model.intercept)

print("Probabilities:", model.pred_proabablity(X))
print("Predictions:", model.predict(X))
```
